```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa['nationality'].value_counts()[0:10]
```

```python
plt.figure(figsize = (15,8))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()),list(fifa['nationality'].value_counts()[0:10]),color = 'g')
plt.suptitle('No.of players across countries',fontweight = 'bold',fontsize = 12)
plt.title('Top 10 countries with most football players',fontweight = 'bold',fontsize = 11)
plt.ylabel('No.of.players',fontweight = 'bold',fontsize = 11)
plt.xlabel('Country',fontweight = 'bold',fontsize = 11)
plt.show()
```

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by = ['wage_eur'],ascending = False)
player_salary.head()
```

Note: We use bar plot whenever we want to obeserve categorical distribution of data.

```python
plt.figure(figsize = (10,8))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color = 'orange')
plt.suptitle('Top 5 highest paid players',fontweight = 'bold',fontsize = 13)
plt.ylabel('Amount in euros',fontweight = 'bold',fontsize=11)
plt.xlabel('Player',fontweight = 'bold',fontsize=11)
plt.text(1,5500000)
plt.show()
```

```python
England = fifa[fifa['nationality']=='England']
England.head(10)
```

```python
England.sort_values(by=['height_cm'],ascending = False).head(3)
```

```python
England_highestpaid_players = England[['short_name','wage_eur']]
England_highestpaid_players.head()
```

```python
England_highestpaid_players=England_highestpaid_players.sort_values(by = ['wage_eur'],ascending = False)
England_highestpaid_players.head()
```

```python
plt.figure(figsize = (10,8))
plt.bar(list(England_highestpaid_players['short_name'][0:10]),list(England_highestpaid_players['wage_eur'][0:10]),color = 'orange')
plt.suptitle('Top 10 highest payed England players',fontsize  = 15, fontweight = 'bold')
plt.ylabel('Salary in Euros',fontsize  = 11,fontweight = 'bold')
plt.xlabel('Player',fontsize = 11,fontweight = 'bold')
plt.title('Here we can oberserve R. Stering is the highest paid football player in England')
plt.show()
```
